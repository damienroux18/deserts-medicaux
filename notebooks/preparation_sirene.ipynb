```python
df2 = (df.loc[df.activiteprincipaleunitelegale.str.startswith(ACTIVITY_CODE)|
              df.activiteprincipaleetablissement.str.startswith(ACTIVITY_CODE)]
        .assign(etab_ape_l1 = lambda x: x.activiteprincipaleetablissement.str.slice(0, 2))
        .assign(el_ape_l1 = lambda x: x.activiteprincipaleunitelegale.str.slice(0, 2))
        .assign(cross_l1 = lambda x: x.el_ape_l1 + "_" + x.etab_ape_l1)
      )
```

```python
# Première Version
df2.loc[df2.cross_l1 == "86_86", 'code_activite'] = df2.activiteprincipaleetablissement
df2.loc[df2.el_ape_l1 != "86", 'code_activite'] = df2.activiteprincipaleetablissement
df2.loc[~df2.activiteprincipaleetablissement.str.startswith(ACTIVITY_CODE, na=False), 'code_activite'] = df2.activiteprincipaleunitelegale
```

```python
#Deuxième Version
df2['code_activite'] = df2.apply(lambda x: x.activiteprincipaleetablissement 
                                  if x.etab_ape_l1 == "86"
                                  else x.activiteprincipaleunitelegale,
                                  axis='columns'
                                 )
```

```python
df2.head(10)
```

```python
#Troisième Version
df2['code_activite'] = df2.activiteprincipaleetablissement
df2['code_activite'] = df2.where(df2.etab_ape_l1 != "86",
                                 df2.activiteprincipaleunitelegale,
                                 axis = 'columns')
```

```python
#Quatrième version
import numpy as np
df2['code_activite'] = np.where(df2.etab_ape_l1 == '86',
                                df2.activiteprincipaleetablissement,
                                df2.activiteprincipaleunitelegale)
```

## Ajout du type de PS & Sauvegarde

```python
df3 =df2.merge(naf_df, how='inner', left_on=['code_activite'], right_on =['NAF'])
```

```python
(df3[['codecommuneetablissement', 'categorie']]
.dropna(subset=['codecommuneetablissement'])
.assign(count = 1)
.groupby(['codecommuneetablissement','categorie'])
.sum()
.reset_index()
.pivot(index='codecommuneetablissement', columns='categorie', values='count')
.fillna(0)
 ).to_csv(os.path.join(PROCESS_DIR, PS_CITIES_FILENAME), sep=';', compression='zip')
```
